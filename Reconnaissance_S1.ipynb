# Séance 1: Reconnaissance et détection

## Import des bibliothèques 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Import et pré-traitement d'image 

**Import et affichage de l'image en couleurs**

```python
image = cv2.imread("image4.jpg")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

**Transformation en niveaux de gris**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.show()
```

**Utilisation de imwrite** 

```python
cv2.imwrite("gray_image.png", gray_image)
```

**Utiliser resize**

```python
new_width = 300
new_height = 200
resized_image = cv2.resize(gray_image, (new_width, new_height))

cv2.imwrite("resized_gray_image.png", resized_image)
```

### Rogner l'image

**Coordonnées du rectangle de la zone à rogner  (x,y) le pixel en haut à gauche**

Pour cibler j'affiche l'image avec les axes (je cible le gateau par exemple)

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
x = 1000
y = 3000
width = 1000  
height = 1500 

cropped_image = image[y:y+height, x:x+width]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  
plt.show()
```

## Application des filtres

### Seuillage

```python
threshold_val = 120
ret, thres_img = cv2.threshold(gray_image, threshold_val, 255, cv2.THRESH_BINARY)

plt.imshow(thres_img, cmap='gray')
plt.axis('off') 
plt.show()
```

### Filtrage

**Flou Gaussien**

```python
image_floue = cv2.GaussianBlur(gray_image, (5,5), 0)
plt.imshow( image_floue,cmap='gray')
plt.axis("off")
plt.show()
```

**Filtre de Canny**

```python
contour_canny = cv2.Canny(image_floue, 30, 40)  

plt.imshow(contour_canny,cmap='gray')
plt.axis("off")
plt.show()
```
